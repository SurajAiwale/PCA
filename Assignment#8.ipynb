# PCA

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import scale
from sklearn.decomposition import PCA
```

```python
df=pd.read_csv("D:\Data Science\Assignments\Assignment #8 PCA\wine.csv")
df
```

```python
df.info()
```

```python
data = df.values
data
```

```python
df_nrml = scale(data)
df_nrml
```

```python
pca = PCA()
pca_var= pd.DataFrame(pca.fit_transform(df_nrml))
pca_var
```

```python
variance = pca.explained_variance_ratio_
variance
```

```python
cumsum = np.cumsum(np.round(variance,decimals=4)*100)
cumsum
```

```python
weight=pd.DataFrame(pca.components_)
sum(weight.iloc[:,9]**2)
```

```python
pca_var.corr()
```

```python
plt.plot(cumsum,color='red')
```

# Hierarchical clustering for PCA

```python
from sklearn.cluster import AgglomerativeClustering
import scipy.cluster.hierarchy as sch
```

```python
hc_df = pd.DataFrame(pca.components_)
hc_df
```

```python
def nrml_fun(i):
    x = (i-i.mean())/(i.std())
    return x
```

```python
nrml_data = nrml_fun(hc_df.iloc[:,0:3])
```

```python
dendrogram = sch.dendrogram(sch.linkage(nrml_data,method='average'))
```

```python
hc = AgglomerativeClustering(n_clusters= 3,affinity='euclidean',linkage='average')
hc
```

```python
hc_y = hc.fit_predict(nrml_data)
hc_y
```

```python
nrml_data['clusterID']= pd.Series(hc_y)
nrml_data
```

```python
hc_result= nrml_data.iloc[:,0:3].groupby(nrml_data.clusterID).mean()
hc_result
```

# K_mean clustering for PCA

```python
from sklearn.cluster import KMeans
from scipy.spatial.distance import cdist
```

```python
k_data = pd.DataFrame(pca.components_)
k_data
```

```python
def k_nrml(i):
    x= (i-i.mean())/(i.std())
    return x
```

```python
k_normal = k_nrml(k_data.iloc[:,0:3])
k_normal
```

```python

k_cluster = KMeans(n_clusters=3)
k_cluster.fit(k_normal)
```

```python
k_cluster.labels_
```

```python
k_cluster.cluster_centers_.shape
k_data['clusterID']= pd.Series(k_cluster.labels_)
k_data
result = k_data.iloc[:,0:3].groupby(k_data.clusterID).mean()
result
```

```python

k= list(range(2,12))
k
TWSS =[]
for i in k:
    kmean = KMeans(n_clusters=i)
    kmean.fit(k_normal)
    WSS= []
    for j in range(i):
        WSS.append(sum(cdist(k_normal.iloc[kmean.labels_==j,:],kmean.cluster_centers_[j].reshape(1,k_normal.shape[1]),"euclidean")))
        
    print("TWSS when k is",i,sum(WSS))
    TWSS.append(sum(WSS))
```

```python
plt.plot(k,TWSS,'ro-')
plt.xlabel('no of cluster')
plt.ylabel('total within sum of squared values')
plt.show
```
